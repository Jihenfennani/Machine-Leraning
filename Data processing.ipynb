La normalisation peut- être effectuée par la technique du Min-Max Scaling.

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Les données doivent être stockées dans un DataFrame Pandas pour être manipulées
data_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/cpu-performance/machine.data'
column_names = ['constructor', 'Model', 'MYCT', 'MMIN', 'MMAX', 'CACH', 'CHMIN', 'CHMAX', 'PRP', 'ERP']
dataset = pd.read_csv(data_url, names=column_names)

columns_to_normalize = ['MYCT', 'MMAX']

subset = dataset[columns_to_normalize]

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(subset)

dataset[columns_to_normalize] = scaled_data

print("Données après normalisation :")
print(dataset)
```

# Partie 4 : Réduction de dimension

### 1- Réduction basée sur une sélection des caractéristiques

#### a- Méthode Filter (Filtre)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_selection import f_classif, chi2, mutual_info_classif
from sklearn.datasets import load_iris

# Charger l'ensemble de données Iris
iris = load_iris()
X, y = iris.data, iris.target

# Calculer les scores de corrélation des caractéristiques
chi2_score, chi2_p_value = chi2(X, y)
f_score, f_p_value = f_classif(X, y)
mut_info_score = mutual_info_classif(X, y)

# Afficher les scores de corrélation des caractéristiques
print("Les scores de corrélation des features :")
print('chi2 score ', chi2_score)
print('F - score score ', f_score)
print('mutual info ', mut_info_score)

# Créer un DataFrame pour visualiser les corrélations entre les caractéristiques
dataframe = pd.DataFrame(iris.data, columns=iris.feature_names)
corr = dataframe.corr()

# Tracer un heatmap pour visualiser les corrélations
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Heatmap des corrélations entre les caractéristiques')
plt.show()
```

#### b-Méthode Wrapper (enveloppante)

```python
import pandas as pd
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les données
iris = load_iris()
X, y = iris.data, iris.target
dataframe = pd.DataFrame(X, columns=iris.feature_names)

# Créer un modèle de régression logistique
model = LogisticRegression(max_iter=1000)

# Créer un objet RFE avec le modèle et le nombre de caractéristiques à sélectionner
rfe = RFE(model, n_features_to_select=2)

# Adapter le RFE au jeu de données
fit = rfe.fit(X, y)

# Afficher les résultats
print("Numéro de caractéristiques sélectionnées:", fit.n_features_)
print("Caractéristiques sélectionnées:", dataframe.columns[fit.support_])

# Afficher les données avec seulement les caractéristiques sélectionnées
selected_features_data = dataframe[dataframe.columns[fit.support_]]
print("\nDonnées avec seulement les caractéristiques sélectionnées:")
print(selected_features_data)

# Afficher la heatmap de corrélation
corr = selected_features_data.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.show()
```

#### c- Méthode Embedded (intégrée)

```python
import pandas as pd
from sklearn.linear_model import LassoCV
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
import seaborn as sns

# Charger les données
iris = load_iris()
X, y = iris.data, iris.target
dataframe = pd.DataFrame(X, columns=iris.feature_names)

# Créer un modèle de régression linéaire LassoCV
lasso_model = LassoCV()

# Adapter le modèle aux données
lasso_model.fit(X, y)

# Obtenir les coefficients des caractéristiques
coefficients = lasso_model.coef_

# Sélectionner les caractéristiques avec des coefficients non nuls
selected_features = dataframe.columns[coefficients != 0]

# Afficher les résultats
print("Caractéristiques sélectionnées:", selected_features)

# Afficher les données avec seulement les caractéristiques sélectionnées
selected_features_data = dataframe[selected_features]
print("\nDonnées avec seulement les caractéristiques sélectionnées:")
print(selected_features_data)

# Afficher la heatmap de corrélation
corr = selected_features_data.corr()
plt.figure(figsize=(8, 6))  # Ajuster la taille de la figure
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")

# Ajuster les étiquettes de l'axe x
plt.xticks(rotation=45, ha="right")  # Rotation de 45 degrés et alignement à droite
plt.show()
```

### 2- Réduction basée sur une transformation des données (PCA)

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

# Charger les données
iris = load_iris()
X, y = iris.data, iris.target
feature_names = iris.feature_names

# Créer un DataFrame à partir des données
df = pd.DataFrame(X, columns=feature_names)
target_names = iris.target_names

# Normaliser les données (important pour PCA)
df_normalized = (df - df.mean()) / df.std()

# Appliquer l'analyse en composantes principales (PCA)
pca = PCA(n_components=2)  # Vous pouvez ajuster le nombre de composantes selon vos besoins
X_pca = pca.fit_transform(df_normalized)

# Créer un DataFrame à partir des nouvelles composantes
df_pca = pd.DataFrame(data=X_pca, columns=['Principal Component 1', 'Principal Component 2'])
df_pca['Target'] = target_names[y]

# Afficher le résultat
plt.figure(figsize=(10, 8))
sns.scatterplot(x='Principal Component 1', y='Principal Component 2', hue='Target', data=df_pca, palette='viridis', s=100)
plt.title('Réduction de dimension avec PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()
```

## Gestion des valeurs manquantes :



```python
# Import des bibliothèques
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA

# Charger les données depuis le fichier CSV
data = pd.read_csv("salary_data.csv")

# 1. Gestion des valeurs manquantes
# Suppression des lignes avec des valeurs manquantes (à adapter en fonction de votre cas)
data.dropna(inplace=True)

# Remplacement des valeurs manquantes par la médiane
median_salary = data['average_salary'].median()
data['average_salary'].fillna(median_salary, inplace=True)

# 2. Encodage des variables catégorielles (exemple avec one-hot encoding)
data_encoded = pd.get_dummies(data, columns=['continent_name'])

# 3. Normalisation (exemple avec Min-Max Scaling)
scaler = MinMaxScaler()
data_normalized = scaler.fit_transform(data[['average_salary']])

# 4. Réduction de dimension (exemple avec PCA)
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data[['median_salary', 'average_salary']])

# Afficher les résultats ou effectuer d'autres opérations
print("Données après prétraitement:")
print(data_encoded.head())
print("\nDonnées normalisées:")
print(data_normalized[:5])
print("\nDonnées après réduction de dimension:")
print(data_pca[:5])
```
